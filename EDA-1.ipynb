```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Task 1: Data Cleaning and Preparation
# Load the dataset
df = pd.read_csv('Cardiotocographic.csv')
```

```python
# Handle missing values
df.fillna(df.mean(), inplace=True)  # Impute missing values with mean
```

```python
# Task 2: Statistical Summary
# Compute measures of central tendency and dispersion
summary_stats = df.describe()
print(summary_stats)
```

```python
# Task 3: Data Visualization
# Create histograms for numerical variables
numerical_vars = df.columns
for var in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[var], kde=True)
    plt.title(f'Histogram of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Create boxplots for numerical variables
for var in numerical_vars:
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[var])
    plt.title(f'Boxplot of {var}')
    plt.ylabel(var)
    plt.show()
```

```python
# Task 4: Pattern Recognition and Insights
# Calculate correlation matrix
correlation_matrix = df.corr()
```

```python
# Visualize correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
